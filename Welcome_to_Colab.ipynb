```
import pandas as pd
import re

# Load the file (make sure you upload Contact_Template_ZeroFixed.xlsx in Colab)
df = pd.read_excel("Contact_Template_ZeroFixed.xlsx")

# Clean ID/Passport column: keep only numbers
if "ID/Passport" in df.columns:
    df["ID/Passport"] = df["ID/Passport"].astype(str)
    df["ID/Passport"] = df["ID/Passport"].apply(lambda x: re.sub(r'\D', '', x))

# Columns that need leading zeros and must be 10 digits
phone_columns = ["Cellphone", "HomeTel"]

for col in phone_columns:
    if col in df.columns:
        df[col] = df[col].astype(str)
        df[col] = df[col].str.replace(r'\D', '', regex=True)  # remove non-numeric
        df[col] = df[col].apply(lambda x: x.zfill(10) if x.strip() != "" else "")

# Display cleaned data
df.head(20)
```

```
output_file = "Cleaned_Contact_Template.xlsx"
df.to_excel(output_file, index=False)
output_file
```
